```python
#we will use the same csv file
import pandas as pd
df = pd.read_csv("/content/pandas_practice_dataset.csv") #loading of data
print(df.head())
print(df.tail())
```

```python
def get_grade(score):
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "F"

df["Grade"] = df["Score"].apply(get_grade)
```

```python
scores = [78, 92, 55, 84]
grades = []

for score in scores:
    grades.append(get_grade(score))

print(grades)  # ['C', 'A', 'F', 'B']
```
